Using a keyword as a variable name will result in a syntax error, because Python expects the keyword to serve its special purpose, not act as a user-defined name.

Example:

Therefore, when naming variables, we must avoid using any of Python's reserved keywords.


**5.What is mutability in Python?**

Mutability in Python refers to the ability of an object to be changed after it is created.If we can change the value or content of an object without creating a new one, it is called a mutable object.
If we cannot change the object after it is created, it is called immutable.

Examples:

Mutable objects: list, set, dictionary

Immutable objects: int, float, string, tuple

**6.Why are lists mutable, but tuples are immutable?**

Ans:-

Lists are mutable because they are designed to store data that may need to be updated or changed. We can add, remove, or change elements in a list after it is created.

On the other hand, tuples are immutable because they are meant to store fixed or constant data. Once a tuple is created, we cannot change, add, or remove its elements.

 **9.What is type casting in Python?**

Ans:-
   Type casting in Python means converting the data type of a value from one type to another. It is used when we want to perform operations between different data types or store values in a specific format.

There are two types of type casting:

1.Implicit Type Casting :-

This is done automatically by Python when it converts a smaller data type to a larger one to avoid data loss.

Example:

x = 5        #int  
y = 2.5       #float  
result = x + y  

In this example, Python automatically converts x (int) to float before addition.

Explicit Type Casting :-
Done manually by the programmer using functions like int(), float(), str(), etc.

Example:

a = "10"

b = int(a)                                        #b is now an integer

In this case, the string "10" is explicitly converted into an integer using
int().

**10.What is the difference between implicit and explicit type casting?**


Ans:-

Implicit type casting (also known as coercion) :- is done automatically by Python. This happens when Python needs to perform an operation on different data types and can safely convert one type to another without losing information. For example, when you add an integer and a float, Python implicitly converts the integer to a float before performing the addition.

Explicit type casting (also known as casting) :- is done manually by the programmer using built-in functions like int(), float(), str(), etc. This is necessary when Python cannot automatically convert the data type or when you want to force a conversion that might result in data loss. For example, converting a float to an integer will truncate the decimal part.

**12. How does the elif statement work?**

Ans:-
The elif statement in Python stands for "else if". It's used in conjunction with the if statement to check multiple conditions sequentially.

how it works:-

1. The if statement's condition is checked first. If it's true, the code block  under the if is executed, and the rest of the elif and else blocks are skipped.

2. If the if condition is false, the program moves to the first elif statement. Its condition is checked. If it's true, the code block under that elif is executed, and the remaining elif and else blocks are skipped.

3. This process continues for each elif statement in order.

4. If none of the if or elif conditions are true, the code block under the else statement (if present) is executed.





**13.What is the difference between for and while loops**

Ans:-

In Python, both for and while loops are used to execute a block of code repeatedly. However, they differ in how they control the repetition.


*  For loops are typically used when you know the number of times you want to iterate or when you want to iterate over a sequence (like a list, tuple, string, or range). They iterate through the items of a sequence or other iterable object.

*  While loops are used when you want to repeat a block of code as long as a certain condition is true. You don't necessarily know in advance how many times the loop will run; it continues until the condition becomes false.

```python
#14.Write a program to reverse a string using a while loop.
name = input("Enter a string: ")
i = len(name) - 1
reverse = ""

while i >= 0:
    reverse = reverse + name[i]
    i = i - 1

print("Reversed string is:", reverse)
```
